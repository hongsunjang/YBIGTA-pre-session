#### groupby를 통해 quality로 묶고 통계값(min,mean,max) 확인

```python
data.groupby(data["quality"]).min()
data.groupby(data["quality"]).mean()
data.groupby(data["quality"]).max()
```

#### groupby를 통해 quality/type로 묶고 개수 세기

```python
data.groupby([data["quality"],data[ "type"]]).count()
```

#### 알코올 함량을 기준으로 오름차순 정리 (20개)

```python
data.sort_values(['alcohol']).head(20)
```

#### 알코올 함량을 기준으로 내림차순 정리 (20개)

```python
data.sort_values(by=['alcohol'], ascending = False).head(20)
```

#### quailty가 7 이상인 데이터 추출 (20개)

```python
data[data['quality'] > 7]
```

#### alcohol을 기준으로 10도가 넘으면 'strong' 아니면 'weak'으로 표시 (lambda 쓰세요)

```python
data.apply(lambda x: 'strong' if (x['alcohol'] > 10) else 'weak', axis = 1).head()
```

#### merge 실습을 위해 데이터를 인위적으로 나눠볼게요

```python
df_x = data[['quality', 'fixed acidity', 'citric acid']]
```

```python
df_y = data[['quality', 'volatile acidity', 'chlorides', 'alcohol']]
```

#### quality를 기준으로 inner join

```python
pd.merge(df_x, df_y, left_on = "quality", right_on = "quality", how = "inner").head()
```

#### quality를 기준으로 outer join

```python
pd.merge(df_x, df_y, left_on= 'quality', right_on = 'quality', how = 'inner').head()
```
